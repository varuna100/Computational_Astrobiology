```python
# libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Reading the file and labelling the columns for RADIANCE SPECTRUM data
df_rad = pd.read_csv('psg_rad.txt', comment='#', delim_whitespace=True) #, header=None)
df_rad.columns = ['Wave/freq', 'Total', 'Noise', 'Stellar', 'GJ-1214b', 'Transit']
```

```python
print (df_rad)  # To display the data and the column variables
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df_rad['Wave/freq'], df_rad['Total'], label='Total Signal', color='black', linewidth=1.5)
#plt.plot(df_rad['Wave/freq'], df_rad['Noise'], label='Noise', alpha=0.7)
#plt.plot(df_rad['Wave/freq'], df_rad['Stellar'], label='Stellar', alpha=0.7) # This one is not necessary
plt.plot(df_rad['Wave/freq'], df_rad['GJ-1214b'], label='GJ-1214b', alpha=0.7)
plt.plot(df_rad['Wave/freq'], df_rad['Transit'], label='Transit', alpha=0.7)



plt.xlabel('Wavelength [$\mu$m]')
plt.xscale('log')
plt.ylabel('Radiance [W/sr/m$^2$/$\mu$m]')
plt.title('Simulated Planetary Spectrum for GJ-1214b - Radiance')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
# Reading the file and labelling the columns for TRANSMITTANCE SPECTRUM data
df_trn = pd.read_csv('psg_trn.txt', comment='#', delim_whitespace=True) #, header=None)
df_trn.columns = ['Wave/freq', 'Total', 'H2', 'He', 'H2O', 'CH4', 'Rayleigh', 'CIA']
```

```python
print(df_trn) #displaying the data in columns for transmittance
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df_trn['Wave/freq'], df_trn['Total'], label='Total Signal', color='black', linewidth=1.5)
plt.plot(df_trn['Wave/freq'], df_trn['H2'], label='H2', alpha=0.7)
plt.plot(df_trn['Wave/freq'], df_trn['He'], label='He', alpha=0.7)
plt.plot(df_trn['Wave/freq'], df_trn['H2O'], label='H2O', alpha=0.7)
plt.plot(df_trn['Wave/freq'], df_trn['CH4'], label='CH4', alpha=0.7)
plt.plot(df_trn['Wave/freq'], df_trn['Rayleigh'], label='Rayleigh', alpha=0.7)
plt.plot(df_trn['Wave/freq'], df_trn['CIA'], label='CIA', alpha=0.7)


plt.xlabel('Wavelength [$\mu$m]')
plt.xscale('log')
plt.ylabel('Transmittance')
plt.title('Simulated Planetary Spectrum for GJ-1214b - Tansmittance')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
